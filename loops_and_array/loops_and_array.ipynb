### Exercise 2.3 — Huber Loss

Huber loss is a hybrid: squared error for small errors, absolute error for large ones.
This makes it robust to outliers.

- If |y - y_pred| <= delta: `0.5 * (y - y_pred)^2`
- Else: `delta * |y - y_pred| - 0.5 * delta^2`

**Worked example:** y_true=[5,2,7], y_pred=[4.8,2.5,10], delta=1
