### Problem
Given `data`, its `mean_value`, its `std_dev`, and a threshold multiplier `k` (commonly 2 or
3), find every value whose distance from the mean exceeds `k * std_dev`.

```
data = [12, 15, 14, 10, 100, 13, 11, 16, 9, 14]
mean_value = 21.4
variance = 690.84
std_dev = variance ** 0.5
k = 2
```

Loop through `data`, and collect the outliers into a list called `outliers`.


````python
data = [12, 15, 14, 10, 100, 13, 11, 16, 9, 14]
mean_value = 21.4
variance = 690.84
std_dev = variance ** 0.5
k = 2

# TODO: loop through data and collect every value whose distance from
# the mean exceeds k * std_dev
outliers = []
````

### Test your solution

Run the cell below. If you see `All tests passed!`, you're done.


````python
expected = [x for x in data if abs(x - mean_value) > k * std_dev]
assert outliers == expected, f"Expected {expected}, got {outliers}"
print("All tests passed! outliers =", outliers)
````
