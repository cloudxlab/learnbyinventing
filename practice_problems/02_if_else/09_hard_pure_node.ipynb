# Decide Whether a Tree Node Is Pure

**Chapter:** If / Else &nbsp;·&nbsp; **Difficulty:** Hard &nbsp;·&nbsp; **ML/Stats concept:** Decision Trees

### Problem
In a decision tree, a node is **pure** if all the samples that reached it belong to one class.
Given the counts of two classes at a node:

- if either count is `0` → `"pure"`
- else if the two counts are equal → `"mixed (balanced)"`
- else → `"mixed"`

```
count_a = 8
count_b = 0
```

Set `node_status` accordingly.


````python
count_a = 8
count_b = 0

# TODO: classify the node as "pure", "mixed (balanced)", or "mixed"
node_status = None
````

### Test your solution

Run the cell below. If you see `All tests passed!`, you're done.


````python
if count_a == 0 or count_b == 0:
    expected = "pure"
elif count_a == count_b:
    expected = "mixed (balanced)"
else:
    expected = "mixed"
assert node_status == expected, f"Expected {expected!r}, got {node_status!r}"
print("All tests passed! node_status =", node_status)
````
